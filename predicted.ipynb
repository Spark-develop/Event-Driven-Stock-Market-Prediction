```python
import pandas as pd
import firebase_admin
from firebase_admin import credentials,firestore
cred = credentials.Certificate("Data/ceptassets-firebase-adminsdk-e4jmf-090eafef1d.json")
default_app = firebase_admin.initialize_app(cred)
```

```python
df = pd.read_pickle('Data/final_df')
df2 = pd.read_pickle('Data/news_final_df')
```

```python
topL = pd.read_csv("topL.csv")
topList=[]
for k in topL['Company']:
    if(len(k)==1):
        topList.append(k)
    else:
        topList.append(k.split()[0])
```

```python
db = firestore.client()
```

```python
for nw1 in topList:
    doc_ref = db.collection(u'Company').document(nw1)
    df41 = df[df['Company'].str.contains(nw1)]
    df42 = df2[df2['Company'].str.contains(nw1)]
    df5 = df41[df41['Analysis'].str.contains('Neutral')]
    df6 = df41[df41['Analysis'].str.contains('Positive')]
    df7 = df41[df41['Analysis'].str.contains('Negative')]
    df51 = df42[df42['Analysis'].str.contains('Neutral')]
    df61 = df42[df42['Analysis'].str.contains('Positive')]
    df71 = df42[df42['Analysis'].str.contains('Negative')]

    doc_ref.set({
        u'company':'%s'%nw1,
        u'Neutral':'%s'%(len(df5)+len(df51)),
        u'Positive':'%s'%(len(df6)+len(df61)),
        u'Negative':'%s'%(len(df7)+len(df71)),
    })
    print('%s   Neutral-%s'%(nw1,(len(df5)+len(df51))))
    print('%s   Positive-%s'%(nw1,(len(df6)+len(df61))))
    print('%s   Negative-%s'%(nw1,(len(df7)+len(df71))))
    print("\n \n")
```

```python
for nw1 in topList:
    n_df1 = df[df['Company'].str.contains(nw1)]
    lki=0
    for lmn in n_df1['Tweets']:
        doc_refer = db.collection(u'Company').document(nw1).collection('Tweets').document("Tweets-%d"%lki)
        doc_refer.set({u'news':lmn})
        lki+=1
```
